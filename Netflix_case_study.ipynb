```python
# Compute IQR
df['duration']=df['duration'].str.replace(['min','season','seasons'],'').astype('float')
Q1 = df['duration'].quantile(0.25)
Q3 = df['duration'].quantile(0.75)
IQR = Q3 - Q1

# Define limits for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df = df[(df['duration'] >= lower_bound) & (df['duration'] <= upper_bound)]
```

# Q1. How has the number of movies released per year changed over the last 20-30 years?

```python
import pandas as pd
```

```python
df=pd.read_csv('/netflix.csv')
```

```python
df
```

```python
df_recent=df[df['release_year']>=(df['release_year'].max()-30)]
release_counts=df_recent.groupby('release_year').size().reset_index(name='count')
plt.figure(figsize=(10,6))
sns.lineplot(x='release_year',y='count',data=release_counts)
plt.title('Number of Movies Released per Year (Last 20-30 Years)')
plt.xlabel('Year')
plt.ylabel('No. of movies')
plt.show()
```

# Q3. What is the best time to launch a TV show?

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
df['month']=df['date_added'].dt.month_name()
tv_shows=df[df['type']=='TV Show']
monthly_tv_shows=tv_shows['month'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=monthly_tv_shows.index,y=monthly_tv_shows.values,palette='viridis')
plt.title('Best time to launch a TV show')
plt.xlabel('Month')
plt.ylabel('No. of TV Shows')
plt.xticks(rotation=90)
plt.show()
```

# Q4. Analysis of actors/directors of different types of shows/movies.

```python
df[['director', 'cast']] = df[['director', 'cast']].fillna("Unknown")
```

```python
df['director_list'] = df['director'].apply(lambda x: x.split(", ") if x != "Unknown" else [])
df['actor_list'] = df['cast'].apply(lambda x: x.split(", ") if x != "Unknown" else [])
```

```python
directors_df = df.explode('director_list')[['type', 'director_list']]
actors_df = df.explode('actor_list')[['type', 'actor_list']]
```

```python
top_directors = directors_df[directors_df['director_list'] != "Unknown"].groupby(['type', 'director_list']).size().reset_index(name='count')
top_directors = top_directors.sort_values(by=['type', 'count'], ascending=[True, False]).groupby('type').head(10)
```

```python
top_actors = actors_df[actors_df['actor_list'] != "Unknown"].groupby(['type', 'actor_list']).size().reset_index(name='count')
top_actors = top_actors.sort_values(by=['type', 'count'], ascending=[True, False]).groupby('type').head(10)
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(y=top_directors['director_list'], x=top_directors['count'], hue=top_directors['type'], palette="coolwarm")
plt.title("Top 10 Directors in Movies vs TV Shows")
plt.xlabel("Number of Shows/Movies")
plt.ylabel("Director")
plt.legend(title="Type")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(y=top_actors['actor_list'], x=top_actors['count'], hue=top_actors['type'], palette="magma")
plt.title("Top 10 Actors in Movies vs TV Shows")
plt.xlabel("Number of Shows/Movies")
plt.ylabel("Actor")
plt.legend(title="Type")
plt.show()
```

# Q5. Does Netflix has more focus on TV Shows than movies in recent years

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
df['year']=df['date_added'].dt.year
df_recent=df[df['year']>=(df['year'].max()-15)]
content_trend=df_recent.groupby(['year','type']).size().reset_index(name='count')
plt.figure(figsize=(10,6))
```

```python
sns.lineplot(data=content_trend, x='year', y='count', hue='type', marker="o", palette=["blue", "red"])
plt.title("Netflix's Focus on TV Shows vs. Movies Over Recent Years")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.legend(title="Type")
plt.grid(True)
plt.show()
```

# Q6. Understanding what content is available in different countries

```python
df['country']=df['country'].fillna("Unknown")
df['country_list']=df['country'].apply(lambda x: x.split(", ") if x != "Unknown" else [])
country_df=df.explode('country_list')[['country_list','type']]
top_countries=country_df['country_list'].value_counts().head(10).index
country_counts=country_df[country_df['country_list'].isin(top_countries)]
plt.figure(figsize=(10,6))
sns.countplot(data=country_counts, x='country_list', hue='type')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Number of Content')
plt.xticks(rotation=90)
plt.legend(title='Type')
plt.show()
```
